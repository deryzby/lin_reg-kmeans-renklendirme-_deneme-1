```python
import matplotlib.pyplot as plt
plt.scatter(bitki_boy_yaprak_alan["Yaprak Alani (Leaf\nArea) (cm2)"], bitki_boy_yaprak_alan["Bitki Boyu (Plant\nLenght) (cm)"], color="purple")
plt.xlabel("Yaprak Alanı")
plt.ylabel("Bitki Boyu")
plt.title("Bitki Boyu-Yaprak Alanı Grafiği")
plt.show()
```

```python
# sci-kit learn kütüphanesini import edelim
# linear regression modelini oluşturalım
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# formül y = ax + b
# a = 4, b = 5
# rastgele sayılar ile modeli eğitelim
# model.fit ([[1], [2], [3], [4]], [5, 9, 13, 17])
```

```python
# modelimizi eğitelim
# x değişkenimiz Yaprak Alani (Leaf\nArea) (cm2), y değişkenimiz Bitki Boyu (Plant\nLenght) (cm)
model.fit (bitki_boy_yaprak_alan[["Yaprak Alani (Leaf\nArea) (cm2)"]], bitki_boy_yaprak_alan[["Bitki Boyu (Plant\nLenght) (cm)"]])
```

```python
tahmin_edilen_degerler = model.predict(bitki_boy_yaprak_alan[["Yaprak Alani (Leaf\nArea) (cm2)"]])
tahmin_edilen_degerler
```

```python
# modelimizin parametrelerini görelim
model.coef_,model.intercept_
```

```python
# predict 1
model.predict([[1]])
```

```python
a = 0.0757296
b = 94.33134525
# Linear Regression Modeli
def Bitki_Boyunu_tahmin_et (Yaprak_Alanı):
    Bitki_Boyu = a * Yaprak_Alanı + b
    return Bitki_Boyu
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(bitki_boy_yaprak_alan[['Yaprak Alani (Leaf\nArea) (cm2)','Bitki Boyu (Plant\nLenght) (cm)']])
bitki_boy_yaprak_alan['küme'] = kmeans.predict(bitki_boy_yaprak_alan[['Yaprak Alani (Leaf\nArea) (cm2)','Bitki Boyu (Plant\nLenght) (cm)']])
bitki_boy_yaprak_alan
```

```python
plt.scatter(bitki_boy_yaprak_alan["Yaprak Alani (Leaf\nArea) (cm2)"], bitki_boy_yaprak_alan["Bitki Boyu (Plant\nLenght) (cm)"], c=bitki_boy_yaprak_alan['küme'])
plt.xlabel("Yaprak Alanı")
plt.ylabel("Bitki Boyu")
plt.title("Bitki Boyuna göre Yaprak Alanı")
plt.show()
```
